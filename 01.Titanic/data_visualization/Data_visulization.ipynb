```python
import os
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('../titanic/train.csv')

# pandas의 read_csv 메소드를 호출하여 train.csv 파일을 읽어오면, dataframe 형태로 반환합니다.
```

```python
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=train)

# hist() 메소를 호출하여 train데이터의 각 컬럼에 대한 분포를 확인한다.
```

```python
sns.barplot(x=train['Sex'], y=train['Survived'], hue=train['Sex'], dodge=False)

# seaborn의 barplot메소드를 호출하여 X축에는 성별, Y축에는 생존비율로 각각 지정하여 성별 별 생존자 분포를 확인한다
#hue = 왼쪽 위의 표를 표시해준다.
```

```python
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=train)
```

```python
plt.subplots(figsize=(10,8))
sns.heatmap(train.corr(), annot=True, linewidths=2)

# seaborn의 heatmap메소드에 train데이터의 상관계수를 인자로 넘겨주면 train데이터의 전체 상관관계에 대한 히트맵을 반환한다.
```

```python
sns.barplot(x='AgeGroup', y='Survived', hue='Pclass', data=train)

# seaborn의 barplot메소드를 호출하여 각 AgeGroup 내의 성별 별 생존률 분포를 확인한다.
```

```python
sns.barplot(x='SibSp', y='Survived', hue='Pclass', data=train, dodge=False)
```
